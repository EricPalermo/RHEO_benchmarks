#### System parameters

```python
#LAMMPS script is in LJ units, so make sure these match

#System size
L = 10 #Size of the system
sf = 0.5 #Unit cell size, relative to L
nb = 1 # # of particles per unit cell
n = nb/(sf**2) # # of particles per area L^2
N = n * L**2

#Fluid properties
eta = 0.01
rho = 1
tau = (rho*(L**2))/(eta*(2*np.pi)**2)
m = rho/n

#Spatial velocity profile
# Of the form u_y = a*sin(bx)
a = 0.1
b = 2*np.pi/L

params = [L,tau]

```

#### Functional Expressions for Velocity

```python
#Timestep in Lammps is a function of the smoothing length
#as per CFL criteria
#h = 3.5*sf
#ts = 0.1*h/4/3

#Array of timesteps
t = np.arange(0,50000, 200)

#Array of x values
x = np.arange(0,L,0.005)

u_x = 0.1*np.sin((2*np.pi*x)/L)
u_t = np.exp(-t/tau)

def sin_x_t(x,t,params):
    u_x = 0.1*np.sin((2*np.pi*x)/params[0])
    u_t = np.exp(-t/params[1])
    return u_x*u_t


#Avg velocity square of each particle
u_avg = ((a**2)/L)*((L/2)-(np.sin(2*b*L)/(4*b)))*np.exp(-2*t/tau)

#Avg kinetic energy per particle
E = 0.5*m*(a**2/L)*((L/2)-(np.sin(2*b*L)/(4*b)))*np.exp(-2*t/tau)

#Total kinetic energy
E_tot = E * n * L**2
```

#### Plot Spatial Velocity Profile

```python
#Obtain velocity and coordinates from the netCDF file at a given timestep nf

def extract_velocities_nf(dataset, nf):
    ds = dataset
    #Arrays of x,y coordinates and velocities at timestep nf
    Xd = ds['coordinates'][nf,:]
    Vd = ds['velocities'][nf,:]

    #Separate arrays for x and y
    xd = Xd[:,0]
    yd = Xd[:,1]

    vyd = Vd[:,0]
    vxd = Vd[:,1]
    
    return xd, vyd
```

```python
# Loop for plotting data from multiple dump files

def plot_dump_files(dump_array, nf, params):
    #Create figure
    fig = plt.figure()

    #Add subplot to figure
    ax = fig.add_subplot(111)
    
    for ds in dump_array:
        xd, vyd = extract_velocities_nf(ds[0],nf)
        
        #Divide the velocity profile into 8 bins
        stat,xe,_ = bstats(xd,vyd,statistic='mean',bins=15)
        xe = (xe[:-1] + xe[1:])/2        
        
        #Plot the results
        ax.plot((xe), (stat), label=ds[1])
        #ax.plot((xd), (vyd), label=ds[1])
           
    #Analytic solution
    #ref_x = np.arange(0,10,0.005)
    #ref_v = sin_x_t(ref_x,nf*200,params)
    #ref_stat,ref_xe,_ = bstats(ref_x,ref_v,statistic='mean',bins=15)
    #ref_xe = (ref_xe[:-1] + ref_xe[1:])/2
    #ax.plot(ref_x,ref_v,label="Analytic Solution")
    
    #Legend and labels
    ax.legend(title = "System Size")
    ax.set_xlabel("x Position")
    ax.set_ylabel("Velocity")
```
